```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
figpath = "/Users/eclin/Desktop/MA Thesis/02 MA Thesis Current Project/Figures/"
```

```python
envdf = pd.read_csv("/Users/eclin/Desktop/MA Thesis/02 MA Thesis Current Project/EnvData/envdata_for_analysis_0504.csv")
```

```python
disaster_type = envdf.groupby("Disaster Type").agg({"Disaster Type" : "count"})
disaster_type = disaster_type.rename(columns={disaster_type.columns[0]: "count_events"})
disaster_type = disaster_type.sort_values(ascending=False, by = "count_events").reset_index()
```

```python
disaster_type
```

```python
figures_path = "/Users/eclin/Desktop/MA Thesis/02 MA Thesis Current Project/Figures/"
cat = disaster_type["Disaster Type"]
vals = disaster_type["count_events"]


plt.figure(figsize=(10,6))
plt.bar(cat, vals, color='lightseagreen', width=0.5)

#create bar plot with custom aesthetics
# topics_gen.plot(kind='bar', title='Number of Bills by Topic Area',
#                ylabel='Number of Bills', xlabel='General Topic Area', figsize=(10, 6),
#                color = "lightsteelblue", textsize = 20)

#rotate x-axis ticks vertically
plt.xticks(rotation=90)

#plt.xlabel('General Topic Area', fontsize = 13)
plt.ylabel('Number of Events', fontsize = 13)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)


plt.savefig(figures_path + "disaster_type_count.jpg", bbox_inches='tight')
plt.show()
```

```python
disaster_subtype = envdf.groupby("Disaster Subtype").agg({"Disaster Subtype" : "count"})
disaster_subtype = disaster_subtype.rename(columns={disaster_subtype.columns[0]: "count_events"})
disaster_subtype = disaster_subtype.sort_values(ascending=False, by = "count_events").reset_index()
```

```python
cat = disaster_subtype["Disaster Subtype"]
vals = disaster_subtype["count_events"]


plt.figure(figsize=(10,6))
plt.bar(cat, vals, color='mediumaquamarine', width=0.5)

#create bar plot with custom aesthetics
# topics_gen.plot(kind='bar', title='Number of Bills by Topic Area',
#                ylabel='Number of Bills', xlabel='General Topic Area', figsize=(10, 6),
#                color = "lightsteelblue", textsize = 20)

#rotate x-axis ticks vertically
plt.xticks(rotation=90)

#plt.xlabel('Disaster-Subtype', fontsize = 13)
plt.ylabel('Number of Events', fontsize = 13)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)


plt.savefig(figures_path + "disaster_subtype.jpg", bbox_inches='tight')
plt.show()
```

```python
n_bins = 30

fig, ax = plt.subplots(1, sharey=True, tight_layout=True)

# We can set the number of bins with the *bins* keyword argument.
ax.hist(envdf["Total Deaths"], bins=n_bins, range=[0, 300],  color = "darkolivegreen")
plt.xlabel('# of Deaths', fontsize = 13)
plt.ylabel('Total Events', fontsize = 13)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

plt.savefig(figpath + "total_deaths_histo.jpeg", bbox_inches='tight')
```

```python
envdf.plot.hist(column = ["Total Deaths"], bins=30, alpha=0.5, range=[0, 100], color = "olive", title='Total Deaths from Env. Event')

plt.savefig(figpath + "total_deaths_histo.jpeg", bbox_inches='tight')
plt.show()
```
